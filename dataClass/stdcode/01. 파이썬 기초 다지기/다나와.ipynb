```python
# 데이터 처리용 라이브러리
import pandas as pd
# 웹페이지 접속 라이브러리
import requests
# 웹페이지소스 이쁘게 수집하는 라이브러리
import bs4
```

```python
page = 1
pageList = []
MAXPAGE = 1000000
```

```python
for page in range(1, MAXPAGE):

    try:
        ### 1. 페이지 접속

        targetUrl = "https://www.ssg.com/search.ssg?target=all&query=%EA%B3%84%EB%9E%80&page={}".format(page)

        # 추가실습 풀 경우 해당 코드로 웹사이트 접근
        user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36'
        resp = requests.get(targetUrl,headers = {'User-Agent':user_agent} )

        bsobj = bs4.BeautifulSoup( resp.text, "html.parser" )

        ### 추가로직 (페이지 없음 처리)
        endResult = bsobj.find(name="div", attrs={"class":"csrch_tip"})
        if ( endResult != None ) :
            break
        else:
            pass

        ### 2. 원하는 데이터 정보 범위 축소

        scope=bsobj.find(name="div",  attrs={"class":"cunit_lst_v"})

        lis = scope.findAll(name="li")

        ### 3. 데이터 수집

        productList = []
        productPriceList = []

        for i in range(0, len(lis)):

            eachProductName = lis[i].find(name="img").attrs["alt"]
            eachProductPrice = lis[i].find(name="em", attrs={"class":"ssg_price"}).text

            productList.append(eachProductName)
            productPriceList.append(eachProductPrice)

        finalDf = pd.DataFrame( zip(productList, productPriceList) )
        pageList.append(finalDf)
        print("페이지 수집완료 : {}".format(page))
    except Exception as e:
        print(e)
```

```python
import requests, bs4
import pandas as pd

def get_egg_data(url, i):
    """
        달걀의 상품 정보를 받아와 csv 파일로 저장하는 함수.
    """
    try:
        # 봇 방지를 위해 추가한 코드
        user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36'
        # url을 통해 웹사이트의 정보를 받아온다.
        resp = requests.get(url,headers = {'User-Agent':user_agent} )
        # 정보를 텍스트 형태로 저장
        html = resp.text
        # 라이브러리를 통해 깔끔하게 정리
        bs = bs4.BeautifulSoup(html, "html.parser")

        ultag = bs.find(name="ul", attrs={"id":"idProductImg"})

        litag = ultag.findAll(name="li")
        # 찾은 데이터 저장을 위한 dic
        egg_data = {
            "name":[],
            "price":[]
        }
        # for문으로 dic 안에 데이터를 저장
        for i in litag:
            info = i.find(name="div", attrs={"class":"cunit_info"})
            egg_data["name"].append(
                info.find(name="em").text
            )
            egg_data["price"].append(
                info.find(name="em", attrs={"class":"ssg_price"}).text
            )

        # 저장된 데이터를 데이터 프레임으로 변환하여 표를 만든다.
        egg_table = pd.DataFrame(zip(egg_data["name"], egg_data["price"]), columns=["상품명", "가격"])
        # 마지막으로 csv 파일으로 저장한다.
        egg_table.to_csv(f"./egg_data{i}.csv", mode="w")
    except EnvironmentError as e:
        print(e)
```

```python
for i in range(1, 10):
    targetUrl = f"https://www.ssg.com/search.ssg?target=all&query=%EA%B3%84%EB%9E%80&page={i}"
    get_egg_data(targetUrl, i)
```
